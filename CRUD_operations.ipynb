### Nested dictionary

```python
#create a nested dictionary with student data
student = {
    "Alice": {"Math": 85, "Science": 92,"English": 95},
    "Bob": {"Math": 78, "Science": 88,"English": 96},
    "Charlie": {"Math": 95, "Science": 90,"English":98}

}
student
```

##For Manipulation using dictionary convert the dictionary to dataframe

```python
import pandas as pd
data = pd.DataFrame(student) #pass the dictionary to data frame and specify an index
data
```

```python
data["Average"] = data.mean(axis=1)
data
```

```python
data["Math"] = data.mean(axis=0)
#axis = 0 denotes columnwise access to the dataframe
data
#The result is not a number
```

```python
import matplotlib.pyplot as plt
#bar chart for student average performance
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color = "skyblue",edgecolor = "black")
plt.title("Average Marks of the students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

##How to visualise subjectwise trend

```python
plt.figure(figsize=(8,5))
# Iterate through the students (columns) to plot their subject scores
for student_name in data.columns[:-2]:  # Exclude 'Average' and the extra 'Math' column
    plt.plot(data.index, data[student_name], marker='o', label=student_name)

plt.title("Average Marks of the students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Re-create the dataframe from the original student dictionary
student = {
    "Alice": {"Math": 85, "Science": 92,"English": 95},
    "Bob": {"Math": 78, "Science": 88,"English": 96},
    "Charlie": {"Math": 95, "Science": 90,"English":98}

}
data = pd.DataFrame(student)

data_transposed = data.transpose() # Transpose the dataframe

plt.figure(figsize=(8,5))
data_transposed[["Math","Science","English"]].plot(kind="box") # Plot box plots for each subject
plt.title("Marks distribution by Subject") # Changed title
plt.xlabel("Subjects") # Changed xlabel
plt.ylabel("Marks") # Changed ylabel
plt.show()
```
